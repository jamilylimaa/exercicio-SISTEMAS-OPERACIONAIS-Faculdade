```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
P = np.array([[0.7, 0.2, 0.1],
              [0.3, 0.5, 0.2],
              [0.4, 0.4, 0.2]])

def simular_compras(P, n):
  probabilidades = np.array([1/3, 1/3, 1/3])
  for _ in range(n):
    probabilidades = np.dot(probabilidades, P)
  return probabilidades
```

```python
n_simulacoes_3 = 2
n_simulacoes_5 = 4

# Simulações
resultados_3a_compra = simular_compras(P, n_simulacoes_3)
resultados_5a_compra = simular_compras(P, n_simulacoes_5)

# Visualização dos resultados
plt.plot(resultados_3a_compra, label="3ª Compra")
plt.plot(resultados_5a_compra, label="5ª Compra")
plt.xlabel("Marcas")
plt.ylabel("Probabilidade")
plt.title("Evolução das Probabilidades de Escolha das Marcas")
plt.legend()
plt.show()
```
